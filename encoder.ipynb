```
import numpy as np
```

```
from sklearn.preprocessing import OneHotEncoder
```

```
encodeur=OneHotEncoder()
```

```
X=[['chat','poil'],['chien','poil'],['poule', 'plume'],['faisan','plume']]
```

```
Xtransform=encodeur.fit_transform(X)
```

```
Xtransform
```

```
Xtransform.toarray()
```

```
Xorigine=encodeur.inverse_transform(Xtransform)
```

```
Xorigine
```
