```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#permet de créer des paquets de pointd appelés blobs !
from sklearn.datasets import make_blobs
```

```python
#on crée 3 clusters artificiels avec une répartition liée à cluster_std
X, y = make_blobs(n_samples=500,centers=3,cluster_std=0.5,random_state=0)
```

```python
sns.scatterplot(X[:,0],X[:,1],style=y) 
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=3)
```

```python
model.fit(X)
```

```python
model.predict(X)
```

```python
sns.scatterplot(X[:,0],X[:,1],hue=model.predict(X)) 
```

```python
model.cluster_centers_
```

```python
model.labels_[:20]#appartenance des 20 premiers points aux différents clusters
```

```python
model.transform(X)[:20] #donne les distances des 20 premiers points par rapport aux centroïds
#pour le coup on voit à quel cluster appartient le point
```

```python
sns.scatterplot(X[:,0],X[:,1],hue=model.predict(X)) 
sns.scatterplot(model.cluster_centers_[:,0],model.cluster_centers_[:,1],marker='o')
plt.title('Clusters avec centroïds')
```

```python
model.inertia_
```

```python
model.score(X)
```

```python
#Méthode Elbow
inertia=[]
K_range=range(1,20)
for k in K_range:
    model=KMeans(n_clusters=k).fit(X)
    inertia.append(model.inertia_)
```

```python
sns.lineplot(K_range,inertia)
plt.title('Méthode Elbow')
plt.xlabel('nombre de clusters')
plt.ylabel('Coût du modèle (Inertia)')
```

```python
from sklearn import metrics
#autre méthode : la méthode silhouette
```

```python
model=KMeans(n_clusters=3)
```

```python
model.fit(X)
```

```python
silhouette_samples=metrics.silhouette_samples(X,model.labels_) #la silhouette mesure les propo
```

```python
silhouette_samples
```

```python
silhouette_score=metrics.silhouette_score(X,model.labels_)
```

```python
silhouette_score  #un bon score, car proche de 1, cela signifie que les points ont été bien mis dans un cluster
#ce score représente la moyenne des silhouette_samples, c'est la myenne des distributions ci-dessous
```

```python
sns.distplot(silhouette_samples) #plus la distribution des distances est proche de 1, mieux c'est
plt.title("distribution des distances des points par rapport à leur centroïds d'adoption")
```

```python
silhouette=[]
for k in range(2,30):
    model=KMeans(n_clusters=k).fit(X)
    silhouette.append(metrics.silhouette_score(X,model.labels_))
```

```python
sns.lineplot(range(2,30),silhouette) #permet de trouver le nombre de cluster optimal quand silhouette est max
#ici c'est donc bien pour n=3
```

```python
 from sklearn.datasets import load_digits
```

```python
digits=load_digits()
images=digits.images
X=digits.data
y=digits.target
```

```python
X.shape
```

```python
y[42]
```

```python
model=IsolationForest(random_state=0,contamination=0.02)
```

```python
model.fit(X)
```

```python
#1 pas anomalie -1 anomalie
model.predict(X)
```

```python
outliers=model.predict(X)==-1
```

```python
outliers
```

```python
plt.imshow(images[outliers][0])
plt.title(y[outliers][0])
```

```python
from sklearn.decomposition import PCA
```

```python
model=PCA(n_components=2)
X_reduced=model.fit_transform(X)
```

```python
sns.scatterplot(X_reduced[:,0],X_reduced[:,1],hue=y,legend='full',palette="Set3")
```

```python
model.components_
```

```python
model=PCA(n_components=64)
X_reduced=model.fit_transform(X)
```

```python
sns.lineplot(range(0,64),np.cumsum(model.explained_variance_ratio_)*100)
```

```python
np.argmax(np.cumsum(model.explained_variance_ratio_) >0.99)
```

```python
model=PCA(n_components=40)
X_reduced=model.fit_transform(X)
```

```python
X_recovered=model.inverse_transform(X_reduced)
```

```python
plt.imshow(X_recovered[0].reshape((8,8)))
```
